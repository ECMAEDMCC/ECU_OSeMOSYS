```python
import pandas as pd
```

```python
output_path = "../f0_OSMOSYS_ECU_Output.csv"
#output_path = "../docs/PLANMICCOutput.csv"
current = pd.read_csv(output_path)
current
```

```python
calibration_data_path = "calibration_data.csv"
calibration_df = pd.read_csv(calibration_data_path)
calibration_df
```

```python
# Years
calibration_years = list(calibration_df['Year'].unique())
```

```python
model_filters = (current['Year'].isin(calibration_years))&(current['Strategy'] == 'BAU') & (current['Emission'] == 'CO2') & (~current['AnnualTechnologyEmission'].isna())
model_df = current.loc[model_filters, ['Year', 'Technology', 'AnnualTechnologyEmission']]
model_df
```

```python
# Calibration 1A1
techs = {
    "1A1": [
"PP_NGSTGS",
"PPINGSPETTST",
"PPINGSPETICE",
"PPINGSPETTGS",
"PP_DSLICE",
"PP_DSLTGS",
"PPIDSLGALICE",
"PPIDSLELRICE",
"PPIDSLELRTST",
"PPIDSLCEMTST",
"PPIDSLPETTGS",
"PPIDSLPETICE",
"PPIFOIPETICE",
"PP_FOIICE",
"PP_FOITST",
"PP_TPP",
"PROCRU",
"REFCRU",
"REFCRUDSL",
"REFCRUGSL",
"REFCRULPG",
"REFCRUFOI",
"REFCRUKJF",
"REFCRUCOK",
"REFCRURED",
"REFNONENE",
"RSVCRU",
"PP_CRU",
"PPICRUPETICE",
"PPICRUPETTST",

    ],
    "1A2": [
        "T5FOIIND",
        "T5LPGIND",
        "T5NGSIND",
        "T5PURDSLIND",
        "T5PURGSLIND",
        "T5COKIND",

    ],
    "1A3": [

        "T4_DSLPUB",
        "T4_GSLPUB",
        "T4_LPGPUB",
        "T4_NGSPUB",
        "T4_DSLPRI",
        "T4_GSLPRI",
        "T4_LPGPRI",
        "T4_NGSPRI",
        "T4_DSLHEA",
        "T4_GSLHEA",
        "T4_LPGHEA",
        "T4_NGSMED",
        "T4_DSLMED",
        "T4_GSLMED",
        "T4_LPGMED",
        "T4_DSLLIG",
        "T4_GSLLIG",
        "T4_LPGLIG",
        "T4_NGSLIG",
        "TRNMIC",
        "TRNBUS",
        "TRNTAX",
        "TRNPASRAI",
        "TRNMOT",
        "TRNSED",
        "TRNSUV",
        "TRNCAM",
        "TRNFREHEA",
        "TRNFREMED",
        "TRNFRELIG",
        "TRNCATTRU",
        "TRNFRERAI",
        "T5KJFAERTRN",
        "T5FOISHITRN",
        "T5PURDSLSHITRN",
        "T5PURGSLSHITRN",

    ],
    "1A4": [
        "T5FOICOM",
        "T5LPGCOM",
        "T5PURGSLCOM",
        "T5PURDSLCOM",
        "T5LPGRES",
        "T5NGSRES",
        "T5LPGAGR",
        "T5PURGSLAGR",
        "T5NOEAGR",
        "T5LPGCON",
        "T5PURGSLCON",
        "T5PURDSLCON",

    ],
    # "1A5": [

    # ]
    "1B2":[
        "T5CRUCRUTRN"
    ]
}

def check_calibration(sub_category:str):
    # Model Results
    model_filters = (model_df['Technology'].isin(techs[sub_category]))
    model_output = model_df.loc[model_filters, ['Year', 'Technology', 'AnnualTechnologyEmission']]
    model = model_output.groupby('Year')[['AnnualTechnologyEmission']].agg('sum').reset_index()
    model['AnnualTechnologyEmission'] = model["AnnualTechnologyEmission"] * 1000 # Mt to Gg
    model = model.rename(columns={"AnnualTechnologyEmission":"Model"})

    # check if all specified techs are found in the model
    model_techs = list(model_output['Technology'].unique())
    if len(model_techs) != len(techs[sub_category]):
        missing_techs = list(set(techs[sub_category]) - set(model_techs))
        print("WARNING: The following techs are missing in the model.")
        print(missing_techs)


    # Calibration Results
    calib = calibration_df.loc[calibration_df['SubCategory']==sub_category, ['Year', 'TotalEmission']]
    calib = calib.rename(columns={"TotalEmission":"INGEI"})

    result = pd.merge(model, calib, on='Year', how='inner')
    result['Error'] = (result['Model'] - result['INGEI'])/ result['INGEI'] * 100
    print(result)
```

```python
print("*** Industrias de la energía (1A1) ***")
check_calibration("1A1")
```

```python
print("*** Industrias manufactureras y de la construcción (1A2) ***")
check_calibration("1A2")
```

```python
print("*** Transporte (1A3) ***")
check_calibration("1A3")
```

```python
print("*** Otros sectores (1A4) ***")
check_calibration("1A4")
```
